```python
dic_country = {}
for country in df_last['location'].unique():
    dic_country[country] = df_last[df_last['location'] == country].copy()
```

```python
dic_country['Germany']
```

 we would like to achieve a mean period in which almost all the countries implemented the quarantine due to COVID-19.

First, we will set the beginning of the pandemic for each country, with the hypothesis that the first non-NaN value in the new_cases column will indicate the approximate date.

```python
begin_pan = {}
for i in dic_country.keys():
    loc = dic_country[i]['new_cases'].first_valid_index()
    if loc is not None:
        if df_last.at[loc,'date'] not in begin_pan.keys():
            begin_pan[df_last.at[loc,'date']] = [1]
        else:
            begin_pan[df_last.at[loc,'date']][0] += 1
    else:
        begin_pan[dic_country[i].date.unique()[0]] = [1]
```

```python
begin_pan
```

```python
pd_begin=pd.DataFrame.from_dict(begin_pan, orient='index', columns=['values'])
```

```python
pd_begin
```

```python
pd_begin.index=pd.to_datetime(pd_begin.index)
```

```python
pd_begin['year'] = pd.DatetimeIndex(pd_begin.index).year
pd_begin['month'] = pd.DatetimeIndex(pd_begin.index).month
pd_begin['day'] = pd.DatetimeIndex(pd_begin.index).day
pd_begin['weekday'] = pd.DatetimeIndex(pd_begin.index).weekday
```

```python
pd_begin
```

It's not possible that the beginning of the COVID-19 Pandemic for any country began in 2021. Also, we want dates before June, since at that point already every country in the world had more than 1 case detected.

```python
begin_df = pd_begin[(pd_begin.year == 2020) & (pd_begin.month <= 6)]
```

```python
begin_df
```

```python
begin_df = begin_df.sort_values(['month','day'], ascending=True)
```

```python
begin_df
```

```python
import plotly.express as px
```

```python
fig = px.histogram(begin_df, x=begin_df.index, y='values', marginal='box')
#fig(figsize=(15,15))
fig.update_layout(title='1. Beginning of Covid 19 around World',
                  xaxis_title='Date',
                  yaxis_title='Count',
                  plot_bgcolor='beige',
                  font=dict(family='Arial', size=14, color='darkblue'))
fig.show()
```

Conclusion :- 
- As we already know, Covid-19 was discovered in December 2019, but it started propagating through the world at the beginning of 2020. It is possible that the first 30 countries to receive the virus between January 19th to February 2nd were from Asia. But we should be careful because when we were counting the beginning dates in each country we needed to include a statement, in which if we can not extract the information required, we just appended the first date we get for that specific country. One example of this situation can be found for Anguilla, which has no data for the first column of the dataset.
- Almost a third part of the world had an increase of new cases in the first half of March, and almost every country in the world had at least one new case by the end of April.
